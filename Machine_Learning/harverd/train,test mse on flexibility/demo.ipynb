```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Ice_cream selling data.csv")
df.shape
```

```python
df.head()
```

```python
plt.plot(df['Temperature (°C)'],df['Ice Cream Sales (units)'],'x')
plt.xlabel('Temp')
plt.ylabel('Ice Cream')
plt.title('Ice Cream Sales vs Temperature')
```

```python
x = df[['Temperature (°C)']].values
y = df[['Ice Cream Sales (units)']].values
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_poly = PolynomialFeatures(degree=2).fit_transform(X_train)
poly_reg = LinearRegression(fit_intercept=False)
poly_reg.fit(x_poly, y_train)
```

```python
y_pred = poly_reg.predict(PolynomialFeatures(degree=2).fit_transform(X_test))
```

```python
plt.scatter(X_test,y_pred)
plt.scatter(x,y)
```
