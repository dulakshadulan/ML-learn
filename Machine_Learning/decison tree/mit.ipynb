```python
import matplotlib.pyplot as plt

# Data points
negative_points = [(-1, -1), (2, 1), (2, -1)]
positive_points = [(-2, 1), (-1, 1), (1, -1)]
green_points = [(1, 1), (1, 2), (-2, -0.01), (-2, 2), (2, 2), (2, -2), (-2, -2)]

# Split into x and y coordinates
neg_x, neg_y = zip(*negative_points)
pos_x, pos_y = zip(*positive_points)
green_x, green_y = zip(*green_points)

# Create the plot
plt.figure(figsize=(8, 8))

# Plot points
plt.scatter(neg_x, neg_y, color='red', label='Negative', s=100)
plt.scatter(pos_x, pos_y, color='blue', label='Positive', s=100)
plt.scatter(green_x, green_y, color='green', label='Test Points', marker='x', s=100)

# Add axis lines
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Grid and labels
plt.grid(True)
plt.legend()
plt.title('Labeled Data and Test Points')
plt.xlabel('f1')
plt.ylabel('f2')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.show()
```
