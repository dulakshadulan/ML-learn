```python
import matplotlib.pyplot as plt 
%matplotlib inline
import numpy as np
import seaborn as sns 

from sklearn.datasets._samples_generator import make_blobs
```

```python
X,y= make_blobs(n_samples=250, n_features=2 ,centers=5 , cluster_std=0.7,random_state=0)
plt.scatter(X[:,0],X[:,1], c=y, s=50 )
plt.title("Scatter plot od Generated Data")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.metrics import pairwise_distances_argmin
def find_clusters(X,n_clusters,rseed = 2):
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]

    while True:
        labels = pairwise_distances_argmin(X,centers)

        new_centers = np.array([X[labels == i].mean(0)for i in range(n_clusters)])

        if np.all(centers == new_centers):
            break
        centers = new_centers
    return centers,labels


    
```

```python
centers,labels = find_clusters(X,5)
plt.scatter(X[:,0],X[:,1],c=labels,s=50,cmap='viridis')
plt.scatter(centers[:,0],centers[:,1],c='black',s=250,alpha=0.5)
```
